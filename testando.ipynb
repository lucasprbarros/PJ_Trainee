```python
import pandas as pd

df_trainee = pd.read_csv(
    r"C:\Users\lupba\OneDrive\Desktop\PJ_Trainee\dados_climaticos_sp_20_25.csv", sep=";", skiprows=9)

df_trainee.head(3)
#lendo o dataset
```

```python
df_trainee.shape
print()
df_trainee.dtypes
print()
df_trainee.info()
print()
df_trainee.describe()
print()
df_trainee.columns
#olhando como tao os dados
```

```python
df_trainee['Data Medicao'] = pd.to_datetime(df_trainee['Data Medicao'])
df_trainee = df_trainee.drop(columns=["Unnamed: 11"])
df_trainee.columns
#tirando a coluna unnamed
```

```python
df_trainee.isnull().sum()
#checando os valores nulos
```

```python
print(df_trainee["Data Medicao"].min())
print(df_trainee["Data Medicao"].max())
#checando as datas
```

```python
#convertendo strings pra float pra poder fazer os gráficos
for coluna in df_trainee.columns[1:]:
    df_trainee[coluna] = (df_trainee[coluna].astype(str).str.replace(",", ".", regex=False))

    df_trainee[coluna] = pd.to_numeric(
        df_trainee[coluna],
        errors="coerce"
    )

df_trainee.dtypes
```

```python
df_trainee.to_csv("dados_climaticos_sp_20_25_limpo.csv", index=False)
```
